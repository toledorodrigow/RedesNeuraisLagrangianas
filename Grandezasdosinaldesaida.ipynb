```python
#Representação de um sistema por meio de sua função de transferência
import numpy as np
from control.matlab import *


#Dados
c = 10e-3
J = 1.3056e-3

s = tf('s')

#Caso1:
zeta1 = 0.7
wn1 = 1.6
Kd = 2*zeta1*wn1* J
Kp = wn1**2 * J
wn1 = np.sqrt(Kp/J)
zeta1 = Kd/(2*J*wn1) 
H1 = (Kp/J)/(s**2 + s*(Kd/J) + Kp/J)
print('\n-----------------------------------------------------\n')
print('H1(s)=',H1)
p1,z1=pzmap(H1)
plt.show()
print('\nresposta ao degrau:')

y1,t1 = step(H1)
fig,ax = plt.subplots(1,1, figsize = (3.5,1.2))
ax.plot(t1,y1) 
ax.set_xlabel(r'Tempo, t[s]',fontsize = 5) 
ax.set_ylabel(r'Resposta em malha fechada, $\theta_p$[rad]',fontsize = 5)

S1 = stepinfo(H1, SettlingTimeThreshold=0.01, RiseTimeLimits=(0.0, 1.0))
print("Coeficiente de amortecimento:",zeta1)
print("frequência natural não amortecida:",wn1)
print('Tempo de subida:', S1["RiseTime"])
print('Tempo de pico:', S1["PeakTime"])
print('Pico:', S1["Peak"])
print('Sobre-sinal (%):', S1["Overshoot"])
print('Tempo de acomodação:', S1["SettlingTime"])
print('Valor de regime permanente:', S1["SteadyStateValue"])
print('\n-----------------------------------------------------\n')

print(wn1)
print(Kp)
print(Kd)
```

```python
#Resposta
import matplotlib.pyplot as plt
from control.matlab import rlocus
from control.matlab import *
from numpy import *
import matplotlib.pyplot as plt
from cmath import polar, phase
from math import degrees
from numpy import conj
from math import *
cm = 1/2.54  # centimeters in inches

s = tf('s')

#Letra A e B: L = (s+10)/(s**2+2s+2)
print("**************LETRA A:L = (s+10)/(s**2+2s+2)**************\n")
L = (s+10)/(s**2+2*s+2)
p,z=pzmap(L)
plt.show()
plt.figure(1,figsize=(30*cm, 15*cm))
rlocus(L)
plt.show()
print("**************FIM LETRA A********************\n")

#Letra C e D


s0 = -20+1j

def H(szero):
  H = (szero+10)/(szero**2+2*szero+2)
  return H

print("==========Calculo dos dados========")
print('H(s0)=',H(s0))
print ('\n arg(s0) =', degrees(phase(s0)), '˚')
print ('\n arg(H(s0)) =', degrees(phase(H(s0))), '˚')
ang = tan(radians(degrees(phase(s0))-degrees(phase(H(s0)))-180))
z = (s0.imag/ang)- s0.real 
print('z=',z)

kp = 1/abs(((s0+z)/s0)*H(s0))
print('kp=',kp)
ki = z/abs(((s0+z)/s0)*H(s0))
print('ki=',ki)
print("============Fim do calculo dos dados==========\n")
print("\n**************LETRA C :L = g(s)h(s)**************\n")

G = (s*kp+ki)/s
H_s = (s+10)/(s**2+2*s+2)
L = G*H_s
print(pole(L))
print(zero(L))
p1,z1=pzmap(L)
plt.show()
rlocus(L)
plt.show()

print("**************FIM********************\n")

print("\n**************LETRA D :y_s=L/(L+1)**************\n")
y_s = L/(1+L)
print(pole(L))
print(zero(L))
p1,z1=pzmap(L)
plt.show()

y1,t1 = step(y_s,T=13)
plt.figure(2)
plt.plot(t1,y1,'g') 
plt.grid(True)
plt.xlabel('Tempo, t') 
plt.ylabel('Saída do sistema, y')
plt.show()
print("**************FIM********************\n")

print("\n**************LETRA D :Ambas juntas**************\n")
y2,t2 = step(H_s,T=13)
plt.figure(3)
plt.plot(t1,y1,'g',t2,y2,'b') 
plt.grid(True)
plt.legend(['ymf(t)','yma(t)'])
plt.title("Saídas no tempo",c='black',fontsize=15)
plt.xlabel('Tempo, t') 
plt.ylabel('Amplitude')
plt.show()
```
